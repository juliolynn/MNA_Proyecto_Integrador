## Ingeniería de características

Copia del texto original convertido a minúsculas

```python
df["lower"] = df["Text"].apply(lambda x: x.lower())
df
```

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, RegexpTokenizer
nltk.download('punkt')
```

```python
tokenizer = RegexpTokenizer(r'\w+')
```

```python
df["tokenized"] = df["lower"].apply(lambda x: tokenizer.tokenize(x))

df.head()
```

Longitud de tokens en cada documento.

```python
df["len"] = df["tokenized"].apply(lambda x: len(x))
df
```

Variable categórica binaria indicando la precencia de la palabra "Shall" en el documento.

```python
df["shall"] = df["lower"].apply(lambda x: 1 if "shall" in x else 0)
df
```

## PCA

```python
from sklearn.feature_extraction.text import TfidfVectorizer 
from sklearn.decomposition import PCA 
```

```python
vectorizer = TfidfVectorizer(stop_words='english')
```

```python
vectorized_documents = vectorizer.fit_transform(df["Text"]) 
```

```python
pca = PCA(n_components=2) 
reduced_data = pca.fit_transform(vectorized_documents.toarray()) 
```

```python
reduced_data
```

```python
df_pca = df.copy()

df_pca['PCA1'] = 0
df_pca['PCA2'] = 0
for i in range(len(reduced_data)):
    df_pca['PCA1'][i] = reduced_data[i][0]
    df_pca['PCA2'][i] = reduced_data[i][1]
```
